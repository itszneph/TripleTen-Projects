```python
# Calculate market share for each genre
def calculate_all_market_shares(sales_data):
    total_market_sales = sales_data['total_sales'].sum()
    market_shares = (sales_data['total_sales'] / total_market_sales) * 100
    return market_shares 

markdown
#VSC-8d373ab9
markdown
<div class="alert alert-success">
  <b>Reviewer’s comment – Iteration 1:</b><br>
  Nice temporal analysis. The grouping and bar chart clearly show the surge in releases after 2000 and the decline after 2010. Your explanation linking this to industry expansion and market corrections makes sense and connects the data to real-world trends.
</div>

markdown
#VSC-0a51553f
markdown
### Key Questions for Genre Analysis:
- Which genres consistently perform well?
- Are there any genres showing recent growth or decline?
- How does the average performance vary across genres?
markdown
#VSC-5d8d8417
markdown
based on the data we can see that the genres action, shooter, sports, and role playing are the majority of the market shares. But when you take into consideration of the amount of releases each genre has some are doing better than others. by checking the amount of game releases per genre compared to the total game releases i compared the percentage of games released to the market share percentage. about 34% of all releases were action yet they hold about 30% of market shares. Role playing had about 13% of all releases being that genre as well as about 13% of the market shares. sports had about 9% of releases and hold about 14% of market shares. shooter on the other hand have about 8% of releases yet hold about 21% of market shares.
markdown
#VSC-e5d1180c
markdown
# Step 4: Regional Market Analysis and User Profiles
markdown
#VSC-8ba9c0fa
markdown
# Identify platforms with declining sales
From the heatmap, many platforms show declining sales over time; this may reflect consumers holding onto existing titles or shifting to new consoles rather than buying every newly released game.
metadata
source
markdown
#VSC-de27a653
markdown
Based on the data, PlayStation platforms (PS3 and PS4) generally performed best across the years, with Xbox platforms (Xbox 360 and Xbox One) often in second place. Most platforms decline after 2014, except for Xbox One and PS4 which maintained stronger sales.
language
source
platform_performance = calculate_region_platform_performance(df_relevant_4)
print(platform_performance.head())

markdown
#VSC-9733a9c5
markdown
### Cross-Regional Platform Comparison
markdown
#VSC-6fcba15a
markdown
I filtered the data for games released on multiple platforms and created a heatmap. Across platforms, total sales for the same title are generally similar; PlayStation and Xbox occasionally deviate from this pattern.
source
code
#VSC-3b26b5c4
python
# Visualize cross-regional comparison for top platforms
platform_performance = platform_performance.set_index('platform')
plt.figure(figsize=(10, 8))
sns.heatmap(platform_performance, annot=True, cmap='YlOrRd', fmt='.1f', cbar_kws={'label': 'Sales (millions USD)'})
plt.title('Video Game Sales by Platform and Region')
plt.xlabel('Region')
plt.ylabel('Platform')
plt.tight_layout()
plt.show()
markdown
#VSC-3d8a5beb
markdown
I filtered for PS4 games and rescaled `user_score` to match the critic score scale. I plotted both scores against total sales and computed correlations: critic score shows a moderate positive correlation (~0.407) with sales, while user score shows little to no correlation (~ -0.032).
source
markdown
#VSC-39b898dc
markdown
## 4.2 Regional Genre Analysis

Now let's examine genre preferences across regions:
code
#VSC-60a875aa
python
# Function to analyze genre performance by region
def calculate_region_genre_performance(data, regions=['na_sales', 'eu_sales', 'jp_sales']):
    genre_totals = data.groupby('genre')[regions].sum().reset_index()
    return genre_totals



markdown
#VSC-04758e03
markdown
### Cross-Regional Genre Comparison

Let's compare genre preferences across regions:
code
#VSC-66ec73f3
python
# Create a comparative genre analysis
genre_performance = calculate_region_genre_performance(df_relevant_4)
print(genre_performance.head())
genre_performance = genre_performance.set_index('genre')
plt.figure(figsize=(10, 8))
sns.heatmap(genre_performance, annot=True, cmap='YlOrRd', fmt='.1f', cbar_kws={'label': 'Sales (millions USD)'})
plt.title('Video Game Sales by Genre and Region')
plt.xlabel('Region')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()


markdown
#VSC-9749abb2
markdown
## 4.3 ESRB Rating Impact Analysis

Finally, let's examine how ESRB ratings affect sales in each region:
code
#VSC-96e0961d
python
# Function to analyze ESRB rating impact
def calculate_region_rating_preformance(data, regions=['na_sales', 'eu_sales', 'jp_sales']):
    rating_average = data.groupby('rating')[regions].mean().reset_index()
    return rating_average



code
#VSC-9a6e8661
python
# Analyze ESRB impact for each region
rating_performance = calculate_region_rating_preformance(df_relevant_4)
print(rating_performance.head())
rating_performance = rating_performance.set_index('rating')
plt.figure(figsize=(6, 4))
sns.heatmap(rating_performance, annot=True, cmap='YlOrRd', fmt='.2f', cbar_kws={'label': 'Sales (millions USD)'})
plt.title('Video Game Sales by Rating and Region')
plt.xlabel('Region')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()

markdown
#VSC-60476ecc
markdown
The genre and regional sales reflect expected market shares. M-rated and E-rated games are popular—likely driven by shooter and sports genres. Note that many games lack ratings, which may affect rating-based analyses.
markdown
#VSC-d80b2302
markdown
# Step 5 : Hypothesis Tests

—Average user ratings of the Xbox One and PC platforms are the same.

—Average user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

—How you formulated the null and alternative hypotheses

—What criteria you used to test the hypotheses~~,~~ and why

code
#VSC-d85336ff
python
games_with_scores = df_relevant_4.dropna(subset = ['user_score'])
xbox_scores = games_with_scores[games_with_scores['platform'] == 'XOne']['user_score']
pc_scores = games_with_scores[games_with_scores['platform'] == 'PC']['user_score']

alpha = 0.01
statistic, p_value = stats.ttest_ind(xbox_scores, pc_scores)
print(f"Alpha level: {alpha}")
print(f"P-value: {p_value}")
print(f"Statistic: {statistic}")

if p_value < alpha:
    print("Reject null hypothesis, user scores are different")
else:
    print("Fail to reject null hypothesis, no significant difference")

action_scores = games_with_scores[games_with_scores['genre'] == 'Action']['user_score']
sports_scores = games_with_scores[games_with_scores['genre'] == 'Sports']['user_score']


alpha = 0.01
statistic, p_value = stats.ttest_ind(action_scores, sports_scores)
print(f"Alpha level: {alpha}")
print(f"P-value: {p_value}")
print(f"Statistic: {statistic}")

if p_value < alpha:
    print("Reject null hypothesis, user scores are different")
else:
    print("Fail to reject null hypothesis, no significant difference")

markdown
#VSC-d89df391
markdown
I formulated the hypotheses and removed missing `user_score` values to ensure valid tests. I selected an alpha of 0.01 for stricter significance. The t-test comparing Xbox One and PC user scores showed no significant difference (t ≈ 1.48). The test comparing Action vs. Sports user scores was significant (t ≈ 11.54), indicating Action has a higher average user score.
markdown
#VSC-bac5bc8f
markdown
# Step 6. Write a general conclusion

markdown
#VSC-8e8973c9
markdown
Based on the analysis, I recommend focusing on an M-rated shooter for PS4 and Xbox One targeting North America and Europe. Between 2013–2016, PS4 and Xbox One generated the highest sales. Shooter games represent about 8% of releases but account for over 20% of market share, and M-rated titles tend to sell best. Critic scores also correlate moderately with sales, so prioritizing quality (as measured by critic reviews) should improve commercial performance.
```
